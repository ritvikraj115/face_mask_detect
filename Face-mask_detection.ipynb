```python
import pandas as pd
import os
import shutil

# Load CSV file
csv_file_path = r'G:\ML\train\_classes.csv'
df = pd.read_csv(csv_file_path)

# Input and output directories
input_image_dir = r'G:\ML\train'
output_mask_dir = r'G:\ML\train\with_mask'
output_no_mask_dir = r'G:\ML\train\without_mask'

# Iterate through CSV rows and move images
for index, row in df.iterrows():
    filename = row['filename']
    label = row.tolist()[1]
    label2 = row.tolist()[2]  #Assuming 'Label' is the column containing mask information
    a= os.path.join(input_image_dir, filename)
    # Check if the image file exists
    if os.path.exists(a):
        # Destination folder based on the label
        destination_folder = output_mask_dir if label == 1 else output_no_mask_dir
        
        # Move the image file to the appropriate folder
        shutil.move(os.path.join(input_image_dir, filename), os.path.join(destination_folder, filename))
    else:
        print(f"Image file {filename} not found!")

print("Classification and moving complete.")
```

```python
def draw_label(img, text,pos, bg_color):
    text_size = cv2.getTextSize(text,cv2.FONT_HERSHEY_SIMPLEX,1,cv2.FILLED)
    font=cv2.FONT_HERSHEY_SIMPLEX
    end_x= pos[0] + text_size[0][0] + 2
    end_y= pos[1] + text_size[0][1] -2

    cv2.rectangle(img,pos,(end_x,end_y),bg_color,cv2.FILLED)
    cv2.putText(img,text,pos,font,1,(0,0,0),1,cv2.LINE_AA)
```
